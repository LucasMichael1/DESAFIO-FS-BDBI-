b.	Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);

```python
df2 = df.groupby('Department')[['EmployeeNumber']].count()
df2.plot(kind='line', marker='4')
plt.suptitle('Quantidade de funcionários para cada nível de escolaridade')
plt.show()
```
